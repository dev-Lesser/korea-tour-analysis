```python
import pandas as pd
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.style.use(['seaborn-whitegrid'])
plt.rcParams['font.family'] = 'Malgun Gothic'
%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_csv('./data/광역지자체 목적지 검색건수_20210621.csv', header=1)
```

```python
new_header = [str(j)+'-'+str(i).zfill(2) for j in range(2019,2021) for i in range(1,13) ]+[str(2021)+'-'+str(i).zfill(2)  for i in range(1,6) ]
```

```python
df.head(2)
```

```python
df.columns = ['광역지자체','구분']+new_header
```

```python
def string_to_numeric(d):
    d = d.replace(',','')
    return int(d)
```

```python
# 숫자 변환
df[new_header] = df[new_header].applymap(string_to_numeric)
```

```python
df['구분'].unique()
```

```python
sido_list = df['광역지자체'].unique()
```

```python
fig, ax = plt.subplots(6,3)
row, col =0,0
for i,isido in enumerate(sido_list):
    df[(df['광역지자체']==isido)&(df['구분']=='자연관광지')].plot.bar(figsize=(15,20), 
                                                     legend=False, 
                                                     title=isido+"=자연관광지 검색건수", 
                                                    ax=ax[row][col], 
                                                    )
    col+=1
    if col==3:
        col=0
        row+=1
    
plt.savefig('./results/a.png', dpi=100)
```
